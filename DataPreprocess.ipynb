# dataset 1

Description: delete duplicated molecules for GCN training

```python
from rdkit import Chem
import pandas as pd
import os
datasetPath = './dataset'
excelName = 'data.xlsx'
csvName = 'dataForGCN.csv'
csvNameGeneral = 'dataForGCNGeneral.csv'
```

```python
dfRaw = pd.read_excel(os.path.join(datasetPath, excelName))
dfRaw
```

```python
# PID to SMILES
def getSmiles(pid):
    mol = Chem.MolFromMolFile(os.path.join(datasetPath, 'Mol', pid+'.mol'))
    return Chem.MolToSmiles(mol)
dfRaw['SMILES'] = dfRaw.apply(lambda x:getSmiles(x['PID']), axis=1)
dfRaw
```

```python
df = dfRaw.copy(deep=True)
df.drop(['Tg_Method', 'Type', 'Method', 'Solvent', 'Temperature1', 'avg(Temperature1)','Time1', 'avg(Time1)', 'Method2', 'min_temp', 'avg(min_temp)', 'max_temp', 'avg(max_temp)', 'Time2', 'avg(Time2)'], axis=1, inplace=True)
df
```

```python
df.drop_duplicates(subset=['PID'], keep='first', inplace=True)
df
```

```python
length = df.shape[0]
split_num = int(length*0.9)
df_1 = df[:split_num]
df_2 = df[split_num:]
df_1
```

```python
df_2
```

```python
df_1 = df_1.sample(frac=1).reset_index(drop=True)
df_1.to_csv(os.path.join(datasetPath, csvName), encoding='utf8', index=False)
df_2.to_csv(os.path.join(datasetPath, csvNameGeneral), encoding='utf8', index=False)
```

# dataset 2

Description: screen molecules synthesized through method 2 for MMGCN training

```python
csvName = 'dataForMMGCN.csv'
csvNameGeneral = 'dataForMMGCNGeneral.csv'
method = 2
solvent = ['DMAc', 'NMP', 'DMF']
```

```python
dfRaw = pd.read_excel(os.path.join(datasetPath, excelName))
dfRaw
```

```python
# method for synthesizing polyimides
if method>0:
    dfRaw = dfRaw[dfRaw['Method']==method].copy(deep=True)
print('num:', dfRaw.shape[0])
```

```python
# PID to smiles
def getSmiles(pid):
    mol = Chem.MolFromMolFile(os.path.join(datasetPath, 'Mol', pid+'.mol'))
    return Chem.MolToSmiles(mol)
dfRaw['SMILES'] = dfRaw.apply(lambda x:getSmiles(x['PID']), axis=1)
dfRaw
```

```python
dfRaw = dfRaw[dfRaw['Tg']<500]
dfRaw = dfRaw[(dfRaw['Temperature1']!=49) & (dfRaw['Temperature1']!=70)]
dfRaw = dfRaw[(dfRaw['Solvent'] == solvent[0]) | (dfRaw['Solvent'] == solvent[1]) | (dfRaw['Solvent'] == solvent[2])]
dfRaw
```

```python
df = dfRaw.copy(deep=True)
df.drop(['Tg_Method', 'Type', 'Method', 'Temperature1', 'Time1', 'min_temp', 'max_temp', 'Time2'], axis=1, inplace=True)
df
```

```python
df.rename({'avg(Temperature1)': 'temperature1',
           'avg(Time1)': 'time1',
           'Method2': 'method2',
           'avg(min_temp)': 'minTemp',
           'avg(max_temp)': 'maxTemp',
           'avg(Time2)': 'time2'}, axis=1, inplace=True)
order = ['PID', 'SMILES', 'Solvent', 'temperature1', 'time1',
         'method2', 'minTemp', 'maxTemp', 'time2', 'Tg']
df = df[order]
df
```

```python
length = df.shape[0]
split_num = int(length*0.9)
df_1 = df[:split_num]
df_2 = df[split_num:]
df_1
```

```python
df_2
```

```python
df_1 = df_1.sample(frac=1).reset_index(drop=True)
df_1.to_csv(os.path.join(datasetPath, csvName), encoding='utf8', index=False)
df_2.to_csv(os.path.join(datasetPath, csvNameGeneral), encoding='utf8', index=False)
```
